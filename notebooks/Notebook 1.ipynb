```python
import gradio as gr
```

```python
import mlflow.sklearn

# Cargar el modelo registrado
modelo_cargado = mlflow.sklearn.load_model("models:/2014_23_features_rfc/4")

```

```python
def execute_model(data):
    # Hacer predicciones
    y_pred = modelo_cargado.predict(data)
    # enviar las predicciones
    return y_pred

demo = gr.Interface(
    fn=execute_model,
    driverId, constructorId, year, round, grid, constructortotalpoints, constructorposition, drivertotalpoints,driverposition,currentraceposition,statusId                  int64
    inputs=["text", "slider"],
    outputs=["text"],
)

demo.launch(share=True)
```
