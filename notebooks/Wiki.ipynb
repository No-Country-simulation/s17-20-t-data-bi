```python
import requests, os
```

```python
# Wikipedia API query string to get the main image on a page
# (partial URL will be added to the end)
query = 'http://en.wikipedia.org/w/api.php?action=query&prop=pageimages&format=json&piprop=original&titles='
```

```python
# get JSON data w/ API and extract image URL
def get_image_url(partial_url):
    try:
        api_res = requests.get(query + partial_url).json()
        first_part = api_res['query']['pages']
        # this is a way around not knowing the article id number
        for key, value in first_part.items():
            if (value['original']['source']):
                data = value['original']['source']
                return data
    except Exception as exc:
        #print(exc)
        #print("Partial URL: " + partial_url)
        data = None
    return data
```

```python
image_urlsdrivers=[]
image_urlscircuits=[]
image_urlsconstructors=[]
```

```python
# loop to get main image url for each page in list

for the_page in Wikiurlsdrivers.collect():
    # get JSON data and extract image URL
    #print(the_page.url[29:])
    the_url = get_image_url(the_page.url[29:])
    #print(the_url)
    if (the_url):        
        image_urlsdrivers.append(the_url)
    else:
        image_urlsdrivers.append("No image file")
    
```

```python
image_urlsconstructors=[]
image_urlscircuits=[]
```

```python
for the_page in Wikiurlscircuits.collect():
    # get JSON data and extract image URL
    the_url = get_image_url(the_page.url[29:])
    if (the_url):        
        image_urlscircuits.append(the_url)
    else:
        image_urlscircuits.append("No image file")
    
for the_page in Wikiurlsconstructors.collect():
    # get JSON data and extract image URL
    the_url = get_image_url(the_page.url[29:])
    if (the_url):
        image_urlsconstructors.append(the_url)
    else:
        image_urlsconstructors.append("No image file")
    
```
