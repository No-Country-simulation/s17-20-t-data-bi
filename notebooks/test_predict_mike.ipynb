```python
import mlflow
from synapse.ml.predict import MLFlowTransformer
```

```python
df = spark.read.format("delta").load(
    "abfss://883c397f-6409-4489-becd-c2232efd4711@onelake.dfs.fabric.microsoft.com/05bfc39e-219f-42de-9013-1ee05672af29/Tables/racesresults"
)
```

```python
model = MLFlowTransformer(
    inputCols=["raceId","year","round","circuitId","driverId","constructorId","grid","statusId","laps"],
    outputCol="position",
    modelName="rfc1",
    modelVersion=1
)
```

```python
df = model.transform(df)

df.write.format('delta').mode("overwrite").save(
    "abfss://883c397f-6409-4489-becd-c2232efd4711@onelake.dfs.fabric.microsoft.com/05bfc39e-219f-42de-9013-1ee05672af29/Tables/PREDICTIONS"
)
```
