```python
def execute_model(driverId, constructorId, year, round, grid, constructortotalpoints, constructorposition, drivertotalpoints,driverposition,currentraceposition,statusId,age):
    # Hacer predicciones
    data = pd.DataFrame([driverId, constructorId, year, round, grid, constructortotalpoints, constructorposition, drivertotalpoints,driverposition,currentraceposition,statusId,age])
    y_pred = modelo_cargado.predict(data)
    # enviar la prediccion
    return y_pred[0]

demo = gr.Interface(
    fn=execute_model,    
    inputs=[gr.tex,"slider"],
    outputs=["text"],
)

demo.launch(share=True)
```
